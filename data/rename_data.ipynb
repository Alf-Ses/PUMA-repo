```python
import os
from pathlib import Path
```

```python
# Path alla cartella contenente le immagini
images_path = Path("images/")

# Verifica e stampa tutti i file nella cartella
print("Contenuti della cartella:")
for file in images_path.glob("*"):
    print(f" - {file.name}")
```

```python
### images folder ###

# Path alla cartella
images_path = Path("images/")

# Cicla su tutti i file
for file in images_path.glob("*.tif"):
    if "training_set_" in file.name:
        new_name = file.name.replace("training_set_", "")
        new_path = file.with_name(new_name)
        file.rename(new_path)
        print(f"Rename: {file.name} → {new_name}")
    else:
        print(f"No changes: {file.name}")
```
